```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#data harga sayur dan buah tahun 2020 (selama covid) berdasarkan bentuk penyajian nya
data=pd.read_csv("Vegetable Prices 2020.csv")
data
```

**Tujuan** saya adalah untuk memprediksi kira kira bentuk penyajian sayur atau buah yang
paling efisien untuk di beli di masa covid

**User** : a completely global

alasannya karena selama COVID-19,banyak konsumen di seluruh dunia bergantung pada produk dengan masa simpan lebih lama seperti kalengan, beku, atau kering. Dengan memfokuskan pada audiens global,bisa memberikan wawasan tentang produk yang paling efisien di berbagai wilayah, berdasarkan bentuk penyajiannya dan ketersediaan global. Permintaan untuk produk yang efisien secara biaya, mudah diakses, dan tahan lama sangat relevan bagi audiens di seluruh dunia.

**Fungsi** :

*  Memberikan informasi untuk pengambilan keputusan yang lebih
baik : membantu konsumen, pengecer, dan pemerintah memilih produk yang paling efisien secara ekonomi dan tahan lama. Misalnya, konsumen dapat memilih bentuk penyajian yang paling hemat biaya, seperti kalengan atau beku, daripada segar yang mungkin lebih cepat rusak dan mahal.
*  Panduan belanja efisien di masa krisis: visualisasi ini berfungsi sebagai panduan bagi masyarakat untuk membeli produk yang memiliki masa simpan lebih lama dan tetap terjangkau. Ini bisa sangat penting untuk keluarga atau individu yang ingin mengurangi frekuensi belanja.



```python
# Mengubah kolom harga menjadi numerik
data['RetailPrice'] = pd.to_numeric(data['RetailPrice'], errors='coerce')

# Menghitung harga rata-rata per unit untuk setiap bentuk penyajian
avg_price_per_form = data.groupby('Form')['RetailPrice'].mean().reset_index()

# Visualisasi menggunakan bar chart
plt.figure(figsize=(8, 6))
color_mapping = {
    'Frozen': 'blue',
    'Fresh': 'green',
    'Dried': 'yellow',
    'Canned': 'gray'
}
sns.barplot(x='Form', y='RetailPrice', data=avg_price_per_form, palette=color_mapping)
plt.title('Rata-rata Harga per Pound Berdasarkan Bentuk Penyajian')
plt.ylabel('Rata-rata Harga per Pound')
plt.xlabel('Bentuk Penyajian')
plt.show()
```

**Hasil menunjukkan bahwa sayuran kalengan lebih murah dan tahan lama, itu mungkin menjadi pilihan yang efisien selama masa pandemi**
